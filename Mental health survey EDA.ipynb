# **Project Summary -**

Mental health is a critical aspect of overall employee well-being and workplace productivity. 2. Objective
The primary objective of this project is to explore the relationships between demographic, occupational, and psychological variables to understand patterns in mental health treatment-seeking behavior. The goal is to assist employers in building effective mental health support systems and reduce the stigma around mental illness in the workplace.

3. Dataset Overview
The dataset used for this analysis is derived from an anonymous global mental health survey conducted among employees across different countries and industries. It includes 27 columns and over 1,200 responses, covering demographic details (age, gender, country), work-related data (self-employment, remote work, number of employees), and mental health indicators (family history, work interference, treatment status, perceived consequences, etc.).

4. Data Wrangling
To ensure the dataset was analysis-ready, extensive data wrangling steps were performed:

Column standardization: All column names were converted to lowercase with underscores.

Gender normalization: Numerous variations of gender entries were consolidated into three categories: Male, Female, and Other.

Age filtering: Unrealistic age entries (below 10 and above 100) were removed.

Missing values handling: Categorical columns were filled with "Unknown" where appropriate.

Encoding: Key categorical features like treatment, family_history, and work_interfere were encoded numerically for correlation analysis.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/content/survey.csv")

# Step 1: Drop irrelevant columns
df.drop(columns=["Timestamp", "comments"], inplace=True)

# Step 2: Standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")

# Step 3: Handle missing values
# Fill NaNs in categorical columns with 'Unknown'
categorical_cols = df.select_dtypes(include='object').columns
df[categorical_cols] = df[categorical_cols].fillna("Unknown")

# Step 4: Clean 'gender' values
def clean_gender(g):
    g = str(g).strip().lower()
    if g in ['male', 'm', 'man', 'cis male']:
        return 'Male'
    elif g in ['female', 'f', 'woman', 'cis female']:
        return 'Female'
    else:
        return 'Other'

df['gender'] = df['gender'].apply(clean_gender)

# Step 5: Remove unrealistic ages
df = df[(df['age'] > 10) & (df['age'] < 100)]

# Step 6: Standardize 'self_employed' and other key fields
df['self_employed'] = df['self_employed'].replace({'Yes': 'Yes', 'No': 'No', 'Unknown': 'Unknown'})

# Step 7: Reset index
df.reset_index(drop=True, inplace=True)

# Final: View cleaned data
print("Final shape:", df.shape)
print(df.head())
````

3️⃣	Handled Missing Values:
Filled missing values in categorical columns with 'Unknown'.
4️⃣	Standardized Gender Column:

#### Chart - 6 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# First, encode categorical features to numeric for correlation
df_encoded = df.copy()

# Encode treatment (Yes=1, No=0)
df_encoded['treatment'] = df_encoded['treatment'].map({'Yes': 1, 'No': 0, 'Unknown': np.nan})

# Encode family history
df_encoded['family_history'] = df_encoded['family_history'].map({'Yes': 1, 'No': 0, 'Unknown': np.nan})

# Encode self_employed
df_encoded['self_employed'] = df_encoded['self_employed'].map({'Yes': 1, 'No': 0, 'Unknown': np.nan})

# Encode work_interfere ordinally
work_interfere_mapping = {'Never': 0, 'Rarely': 1, 'Sometimes': 2, 'Often': 3, 'Unknown': np.nan}
df_encoded['work_interfere'] = df_encoded['work_interfere'].map(work_interfere_mapping)

# Select numeric columns
corr_features = ['age', 'treatment', 'family_history', 'self_employed', 'work_interfere']

# Compute correlation matrix
corr_matrix = df_encoded[corr_features].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
````

#### Chart - 7 - Pair Plot

````python
# Pair Plot visualization code
# Drop NaNs to avoid issues in pairplot
pair_data = df_encoded[corr_features].dropna()

# Plot pairwise relationships
sns.pairplot(pair_data, diag_kind='kde', corner=True)
plt.suptitle("Pair Plot of Key Features", y=1.02)
plt.show()
````
